## R - CNN

기존의 CNN은 window로 각 픽셀의 위치를 돌아가면서 인식하여 많은 시간이 소요가 되었다면,
R- CNN은 특정 부분들을 크게 잡아 (region propsals)들을 파악하여 시간효율성을 
극대화 하였다 

2000개의 추출된 부분들은 상자안에 분류되어 convolutional 뉴럴 네트워크로 전달이 된다. (벡터4096차원으로 ) 

그 이후에는, 이 전달된 것들이 서포드 벡터 머신(분류)로 적용이 된다. 
이 과정에 있어서 바운딩 박스에 있는 것들을 예측을 할 수 있어, 효율적으로 인식이 가능하다.

개선되여야 할점,
2000개로 나누어진 것들이 아직까지는 시간이 많이 소모 된다. 
고정이 된 부분들이라서 랜덤이나 개선되는 파라미터등들이 추가 될 수 없어 그냥 그대로로 머물어 있다는 것이다. 

## Fast CNN

그래서 위의 R-CNN에서 개선 된 것이 Fast R-CNN 이다. 

비슷한 맥락이지만, generate convolitional feature map 으로 다음단계로 보낸다. 이 부분에서 우리는 RoI pooling layer 을 통해 바운더리로 묶을 수 있다. 
classification으로는 softmax 를 이용하여 예측하여 분류 한다.

이것이 조금더 빠른 이유는 단지 갯수의 차이일 뿐이다. 

+ 여기서 추가적으로 설명할 것 
Region of interest pooling (also known as RoI pooling)


## YOLO

전에서 봤던 것들은 모두 localize를 했던 방식(input이 불안정한 이미지 )이라면, 이 방식은 완전히 다른 방법을 취한다. 
"You Only Look once "

YOLO는 single convolutional network 으로 바운딩박스를 예측하여 분류하는 방식이다.  시간 효율성이 좋은 것 을 알 수 있다.
그림을 보면 이해를 할 수 있다. 

YOLO의 방식을 보면, image를 SxS로 grid를 한다.
각각 하나의 grid된 박스를 보면 박스에는 class probaility 로 인식 및 분류가 가능하다. 하지만, 새의 깃털같은 세밀한 분류는 어려운 편이다.
