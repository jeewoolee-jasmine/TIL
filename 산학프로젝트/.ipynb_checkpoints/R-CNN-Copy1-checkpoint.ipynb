## R - CNN

기존의 CNN은 window로 각 픽셀의 위치를 돌아가면서 인식하여 많은 시간이 소요가 되었다면,
R- CNN은 특정 부분들을 크게 잡아 (region propsals)들을 파악하여 시간효율성을 
극대화 하였다 

## Process 
1. Input image
2. Extract region poprslas(Selective search)
3. 각각의 class마다 SVM으로 classification을 수행
4. CNN features 을 이용하여 bounding box regression model을 수정한다. (localization의 에러를 최소화하기 위하여)

region propsals 의 Selective search 알고리즘 방법  Bottoms - up 방법 

약 2000개의 추출된 부분들은 상자안에 분류되어 convolutional 뉴럴 네트워크로 전달이 된다. (벡터4096차원으로 ) 

그 이후에는, 이 전달된 것들이 서포드 벡터 머신(분류)로 적용이 된다. 
이 과정에 있어서 바운딩 박스에 있는 것들을 예측을 할 수 있어, 효율적으로 인식이 가능하다.

개선되여야 할점,
2000개로 나누어진 것들이 아직까지는 시간이 많이 소모 된다. 
고정이 된 부분들이라서 랜덤이나 개선되는 파라미터등들이 추가 될 수 없어 그냥 그대로로 머물어 있다는 것이다. 

## Fast CNN

## Process 
1. Input image
2. Extract region poprslas(Selective search)
3. input된 image를 CNN으로 여과하여 feature map 을 추출한다. 
4. selective search 된 ROI를 feature map크기에 맞추어 projection 시킨다. 
5. projection 된 ROI에 pooling을 진행하여 Vetor화 한다. 

6.  각각의 class마다 SVM으로 classification을 수행
7. CNN features 을 이용하여 bounding box regression model을 수정한다. (localization의 에러를 최소화하기 위하여)

그래서 위의 R-CNN에서 개선 된 것이 Fast R-CNN 이다. 

비슷한 맥락이지만, generate convolitional feature map 으로 다음단계로 보낸다. 이 부분에서 우리는 RoI pooling layer 을 통해 바운더리로 묶을 수 있다. 
classification으로는 softmax 를 이용하여 예측하여 분류 한다.

이것이 조금더 빠른 이유는  FC layer의 input 부분만 고정이고, 앞의 과정을 다르게 했기 때문이다.

+ 여기서 추가적으로 설명할 것 
Region of interest pooling (also known as RoI pooling)

큰 하나의 바운딩박스를 분해해 분석하는 것으로 볼 수 있다.


마지막으로, pipline으로 연산이 되었던 (SVM과Box regression) 기존의 R-CNN과 달리 
end- to- end로 한번에 학습할 수 있다는 점이 다르다. 

# Faster R- CNN

기존의 fast R-CNN 에서 RPN 네트워크 구조가 추가된 것으로 볼 수있다.

## Process 
1. Input image
2. input된 image를 CNN으로 여과하여 feature map 을 추출한다. 
3. Region Proposal Network 를 여과하여 object를 구별하는 최소한의 proposal의  구분
5.  ROI에 pooling을 진행하여 Vetor화 한다. 
6. SVM으로 classification,  bounding box regression model을 이행한다.

RPN(Region Proposal Network)
RPN의 input 값은 이전 CNN 모델에서 나온 feature map 이다. 
featuremap 에서 sliding window를 진행한다. 

이렇게 추출된 anchor box를 이용하여 classification 과 regression을 먼저 구한다.

그렇게 나온 결과들은 
Postive와 Negative로 나눌수 있다. 
IoU 가 0.7 보다 크거나
anchor box 중 IoU가 가장 크거나

-> Postive (Object로 판별)

IoU가 0.3 보다 작으면 background  로 판별하여

-> Negative

하나의 object 당 여러 개의 proposal 값들이 나오게 되는데,

NMS 알고리즘을 사용하여 개숫를 줄여간다.

1. box들의 Cofidence score 기준으로 정렬한다. 
2. 순차적으로 계산하여 IoU가 0.7 이상이면, 같은 객체를 판별한거라 가정하고
 해당 box는 제거한다. 
 
3. 그래서 max pooling 과 같은 가장 높은 score을 가진 box 들만 남게된다. 

+ RoI Poiling을 진행 할때, 이미지사이즈가 모두 같은 크기로 변경되어 정규화 되어
input image의 사이즈에 영향을받지 않았지만, 각각의 알고리즘 상에 최적화된 규격이 있어서 구현을 할때 사이즈는 고정되어 있다. 

추가적으로, Global Average Pooling 을 사용하면 input size 에 관계없이 filter의 갯수 만 
고정되어 있기 때문에 많이 쓰는 추세이다.

# Mask R-CNN

Faster R-CNN 에서 mask branch 와 FPN, RoI align이 추가된 형태

FPN

Upsampling 하여 (마지막 layer에서의  feature map에서 전의 feature map을 더하는 과정으로)  이전 정보를 저장하여 유지 할 수 있다. 

이렇게 되면, anchor 생성하고 비교하는일 이 없어졌다.

RoI
기존의 방식은 소수점 좌표를 가지면 반올림식으로 이동하여 위치가 왜곡되기 때문에 
bilinear interpolation 이용하여 위치정보를 담는 RoI align 방식을 이용한다.
