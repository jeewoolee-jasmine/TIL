## Label 과 feature 값 나누기 

```python
# 4. feature/label 나누기
x = DF.drop('Salary', axis=1)
x
```

```python
y= DF['Salary']
y
```

```python
#방법2(indsx로 접근) iloc = index location 
x_2 = DF.iloc[:,0:3] # [:, :-1]
print(x_2)
print("===" *9)
y_2 = DF.iloc[:,3]
print(y_2)
```

```python
#비어 있는 값 확인하기 (데이터 전처리용)
x.isnull().sum()

# Age 와 Year 은 Numeric 값으로 숫자이기때문에, 평균값으로 넣어 줄 수 있음. 
```

```python
# 5. 비어있는 값 채우기(mean
```

```python
from sklearn.impute import SimpleImputer
# 비어 있는 imputer 생성 
mean_imputer = SimpleImputer(strategy = 'mean')

# update calcultin 
mean_imputer.fit(x.iloc[:, 1:])

#Data transform (변환하기)
x.iloc[:, 1:]= mean_imputer.transform(x.iloc[:, 1:])
```

```python
x.isnull().sum()
```

```python
# 6. One hot encoding 
#데이터 카테고리형(명목형)을 숫자로 변환시켜주기 
# one hot encoding 은 True False 로 하여 -> Numeric한 value가 아닌 것으로 인식 시켜줘야 한다. 
# array 형태로 반환 해준다. 

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct = ColumnTransformer([('one_hot_encoder', OneHotEncoder(), [0])], remainder = 'passthrough')

x= ct.fit_transform(x)
x


#Encoder의 종류에도 굉장히 많으니 참고
```

```python
# 7. Split Data    # x = feature, y = label
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)

print(X_train)
print('=====' *15)
print(y_train)


# 여기서 split 할때는 random 으로 되기 때문에 random seed 를 지정해야한다. 
# X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
# 8. Train


from sklearn.linear_model import LinearRegression

linear_model= LinearRegression()   # y= wx +b / bias,(기본값이 True : bias의 영향을 미치게 한다. ) intercept
# linear_model =  LinearRegression(intercept = False)  -> 영향을 없애준다. 
linear_model.fit(X_train, y_train)
```

```python
# 8-1. Train_2
from sklearn.tree import DecisionTreeRegressor
tree_model =DecisionTreeRegressor().fit(X_train, y_train)
```

```python
# 9. Score

y_pred = linear_model.predict(X_test)
print(y_pred)
print(y_test)
```

```python
y_pred_tree = tree_model.predict(X_test)
```

```python
# 3-3. 데이터 설명보기
DF.describe()

#아래의 Salary 의 평균값은 63500 아며,
 밑의 Evalute 에서 MAE 값을 비교 해 볼 수있다. 
```

```python
# 10. Evalute
# 현재 하고 있는 학습 = Regression 이니, 그것에 맞는것을 가져와야함. 

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score


mae_l= mean_absolute_error(y_test, y_pred)
sme_l = mean_squared_error(y_test, y_pred)
r2_l = r2_score(y_test, y_pred)

print('mae =', mae_l)
print('msme =', sme_l)
print('r2 =', r2_l)
```

```python
mae_t= mean_absolute_error(y_test, y_pred_tree)
sme_t = mean_squared_error(y_test, y_pred_tree)
r2_t = r2_score(y_test, y_pred_tree)

print('mae =', mae_t)
print('msme =', sme_t)
print('r2 =', r2_t)
```

 실제와 예측 값의 차이가 작을수록 예측률 이 높다. 
 
 하지만, 값이 - 가 나올 수 있다. 그래서 절대값으로 해준다.  제곱후 루트
 

RMSE 는 특정 값들의 영향을 많이 받는다. 그래서 숫자가 커지는 경향이 있는데, 이걸 줄이도록 모델을 다시 만들거나 해야한다.

또한 데이터의 단위에 따라서 같은 값(숫자) 예를들어2의 의미는 상당히 달라진다. 
데이터의 크기가 큰 경우 2이면 정확한 수준이지만, 10단위 밑으로 적은 양이면 오류가 크다고 볼 수있다
