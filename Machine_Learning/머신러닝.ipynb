# 머신러닝 

분류할때 사용한 기준을 하드코딩한다. 
기존의 단순한 크기 비교, If size >7 ....

알아서 기준점을 잡아 분류하는 것이 머신러닝. (기준점은 분류의 정확성을 위해서 유연하게 변경한다. )

기준을 알아서 잡아가는 과정을 "학습" 이라고한다. 

프로그래머가 직접 기준을 입력하는 것을  절차적 프로그램이라고 한다. 

Model: 구별하는 객체 
traingset: 기준선 찾는 것 
Deep learning 여러번 반복 기준 찾음
learning, traing : 학습
test set, evaluation set : 정확도 계산

선형 회귀 

아래와 같은 식을 찾는 알고리즘
1. 최소 자승법 (통계 선형회귀)
2. Gradient Decent  (Deep learning)


기준선에서 x1, x2 계산 값에서 오류가 나온다면, 가중치를 부여해서 어느 기준을 좀 거 세게 
줄 것인지 에 대해서, w1,w2의 값을 조절하여 그래프의 기울기를 조절한다. 

학습을 한다는 건, 선을 구성하는 3개의 값을 찾는 것이다. w1,w2, b의 값, b는 기울기라고 볼 수있움
