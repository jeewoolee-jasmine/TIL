```python
train=pd.read_csv("train.csv.gz")
train
```

```python
test=pd.read_csv("test.csv.gz")
test
```

```python
alldata=pd.concat([train, test]) #병합하기
alldata
```

```python
alldata['DateTime']=pd.to_datetime(alldata['DateTime'])
alldata['Year']=alldata['DateTime'].dt.year
alldata['Month']=alldata['DateTime'].dt.month
alldata['Week']=alldata['DateTime'].dt.week
alldata['WeekDay']=alldata['DateTime'].dt.weekday
alldata['Day']=alldata['DateTime'].dt.day
alldata['Hour']=alldata['DateTime'].dt.hour
alldata['Minute']=alldata['DateTime'].dt.minute
```

```python
alldata2=alldata.drop(['AnimalID','DateTime','OutcomeType','OutcomeSubtype','ID'], axis=1)
alldata2           
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
c=alldata2.columns[alldata2.dtypes==object]
for i in c:
    alldata2[i]=le.fit_transform(list(alldata2[i]))     # list는 null겂이 카테고리형 인식방법으로 가는 것을 방지한다. 
alldata2
```

```python
alldata2.isnull().sum()
```

```python
train2=alldata2[:len(train)]
test2=alldata2[len(train):]
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_jobs=-1)
rf.fit(train2,train['OutcomeType'])
result=rf.predict_proba(test2) #불확실성 추정 방식
```
