```python
#  ,    
s= "pPoooyY"
def solution(s):   
    p_num=s.count('p')     # s   p   
    P_num = s.count('P')   # s   P   
    y_num=s.count('y')      # s   y   
    Y_num=s.count('Y')      # s  Y  
    Total_p = p_num +P_num              #   p,P  
    Total_y = y_num +Y_num              #   y,Y  
    if Total_p != Total_y:
        ans=False
    elif Total_p == Total_y:
        ans=True
    elif Total_p==0 and Total_y==0:
        ans=True
    return  ans
```
