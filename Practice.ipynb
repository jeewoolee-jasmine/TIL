```python
# 문제의 포인트는, 대소문자 구분하여 셀렉하기 
s= "pPoooyY"
def solution(s):   
    p_num=s.count('p')     #문자열 s 중 소문자 p 의 갯수 세기
    P_num = s.count('P')   #문자열 s 중 대문자 P 의 갯수 세기
    y_num=s.count('y')      #문자열 s 중 소문자 y 의 갯수 세기
    Y_num=s.count('Y')      #문자열 s 중 대문자Y의 갯수 세기
    Total_p = p_num +P_num              # 문자열 중 p,P의 갯수 
    Total_y = y_num +Y_num              # 문자열 중 y,Y의 갯수 
    if Total_p != Total_y:
        ans=False
    elif Total_p == Total_y:
        ans=True
    elif Total_p==0 and Total_y==0:
        ans=True
    return  ans
```
