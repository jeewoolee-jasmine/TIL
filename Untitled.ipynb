```python
from pymongo import MongoClient
from gridfs import GridFS
from gridfs import GridFSBucket
```

```python
from PIL import Image
import urllib.request
from io import BytesIO
```

```python
db =MongoClient().python_test
```

```python
fs= GridFS(db)
```

```python
image_list = list(db.fs.files.find())
len(image_list)
```

```python
images=[]
labels=[]
#IM_WIDTH =200,IM_HEIGHT=200 지정해줘도 되고, 함수안에 넣어도 되고 
import numpy as np
```

```python
for file_detail in image_list:
    file_name = file_detail["filename"]   #filename 부분만 추가 원함
    print(file_name)
    
    f=fs.get_last_version(filename=file_name)
    data=f.read()
    
    if len(data)>0:        # 파일이 에러가 아니라면 길이는 0보다 큰 것을 전제조건하에, 
                           # 파일을 이미지화 하기, 흑백으로하기, 크기 맞추기 
        im = Image.open(BytesIO(data))
        im = im.convert('L')
        im = im.resize((200,200)) 
        
        im = np.array(im).flatten()/255   # 0-1 사이의 숫자로, 1줄로 변경해주기
        im= im.astype("float32")   # 머신러닝은 float 형태를 선호하여 변경해주기
        images.append(im)
        label= 1 if "cat" in file_detail["metadata"]["type"] else 0
        labels.append(label)
            
        
```

# 위의 자료를 numpy 배열로 변환

```python
len(images)
```

```python
len(labels)
```

```python
images_array_in_numpy = np.array(images)
labels_array_in_numpy = np.array(labels)
```
