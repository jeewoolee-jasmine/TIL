# 2가지의 데이터가 있다. 1.train 2.test 1번자료를  토대로, 머신러닝의 학습을 유도하여, 2번 데이터의 결과를 예측해보자.

```python
train_file = pd.read_csv('train.csv')
train_file
```

```python
test_file=pd.read_csv('test.csv')
test_file
```

# 두 자료 모두, 결측지 제거를 위해 하나의 데이터(alldata)를 만들어준다.

```python
alldata=pd.concat([train_file, test_file]) #concat = concatenate 의 약자로 together의 의미가 있다.
alldata
```

```python
# 년도를 분리하는 이유는, 데이터 분석시에 영향이 크게 미칠것이라 판단하였기 때문
# openDate의 타입을 데이트타입으로 변경하는 과정 
```

```python
alldata['openDate']= alldata['openDate'].astype(str)   #합병된 데이터의 openDate칼람의 모든 것을 string으로 변환
alldata['openDate']=alldata['openDate'].apply(lambda x: x.split('.')[0])  #변환된것들 중 . 기준으로 2개로 나눔, 그중에 0번째 인덱스
alldata['openDate']=pd.to_datetime(alldata['openDate']) #분리된 숫자들을 날짜형으로 변환
alldata['year']= alldata['openDate'].dt.year  # 그중에 필요한 부분 "연도"로 우선 분류
```

# 여기서부터 데이터의 정형화와 처리과정
