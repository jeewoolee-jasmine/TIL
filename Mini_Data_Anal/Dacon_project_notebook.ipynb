```python
#employee 칼럼 전처리
a=test_file['employee1'].astype(str).tolist()
a2=test_file['employee2'].astype(str).tolist()
b=[]
b2=[]
for i in a:
    if ',' in i:
        i=i.replace(',','')
        b.append(i)
    else:
        b.append(i)
for i in a2:
    if ',' in i:
        i=i.replace(',','')
        b2.append(i)
    else:
        b2.append(i)
b=list(map(float, b))
b2=list(map(float, b2))
test_file['employee1']=b
test_file['employee2']=b2
```
